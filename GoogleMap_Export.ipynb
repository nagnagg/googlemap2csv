```python
# 基準になる位置情報を検索 
geocode_result = client.geocode('多摩境　公園') 
geocode_result
```

```python
#半径〇〇以内の情報を取得
place_results = client.places_nearby(location=loc, radius=100000, keyword='多摩境　公園',language='ja')
pprint.pprint(place_results)
```
